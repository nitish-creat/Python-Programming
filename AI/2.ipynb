```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    "month" : ["jan","feb","mar","apr","may","jun"],
    "productA" : [20000 , 22000, 25000, 27000, 30000, 32000],
    "productB" : [15000 , 18000, 21000, 24000, 27000, 30000],
    "productC" : [10000 , 12000, 15000, 18000, 21000, 24000]
}

df = pd.DataFrame(data)
print(df)
```

```python
df['total_sales'] = df['productA'] + df['productB'] + df['productC']
df['Avg_sales'] = (df['total_sales'] / 3).round(2)
print(df)
```

```python
max_sales = df.loc[df['total_sales'].idxmax()]
print("Month with highest sales:", max_sales['month'])
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['month'], df['productA'], marker='o', label='Product A')
plt.plot(df['month'], df['productB'], marker='o', label='Product B ')
plt.plot(df['month'], df['productC'], marker='o', label='Product C')
plt.title('Monthly Sales of Products A, B, and C')  
plt.show()
```

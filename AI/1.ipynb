```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'name': ['Rajesh', 'Priya', 'Arjun', 'Neha', 'Vikram', 'Ananya'],
    'gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'math': [85, 92, 78, 88, 95, 80],
    'science': [80, 88, 82, 90, 92, 85],
    'english': [75, 95, 70, 92, 88, 90],
    'attendance': [95, 98, 92, 96, 94, 97]
}
```

```python
df =  pd.DataFrame(data)
df['Total'] = df[['math','science','english']].sum(axis=1)
df['Percentage'] = df['Total']/3

print(df)
```

```python
print(f"Average Percentage: {df['Percentage'].mean():.2f}")
print(f"{df.loc[df['Total'].idxmin()]}")

print(f"Average pecentage by gender : {df.groupby('gender')['Percentage'].mean()}")
print("Top 3 Students by Total Score:")
print(df.nlargest(3, 'Total')[['name', 'Total', 'Percentage']])
```

```python
plt.figure(figsize=(10,5))
df[['math','english','science']].mean().plot(kind='bar')
plt.title("Avg marks per subject")
plt.ylabel("Marks")
plt.xlabel("Subject")
plt.show()
```

```python
plt.scatter(df['attendance'],df['Percentage'])
plt.xlabel("Attendence")
plt.ylabel("Percentage")
plt.title("Att vs per")
plt.show()
```

```python
df['gender'].value_counts().plot(kind='pie')
plt.title('Gender Distribution ')
```

```python
plt.hist(df['Percentage'])
plt.title("Percentage Distribution")
plt.xlabel("Percentage")
plt.ylabel("Frequency")
plt.show()
```

```python
df[['math','english','science']].boxplot()
plt.title("Marks dis acrros sub")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[['math','science','english','attendance','Percentage','Total']].corr(), annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
corr = df['attendance'].corr(df['Percentage'])
print(corr)

print(df[['math','science','english']].mean().idxmax())
print(df[['math','science','english']].mean().idxmin())
```

## VIF (Variance Inflation Factor)

why do we use VIF
- to check multicollinearity between independent variables.
- used before regression analysis.

why we use this
- example -> Are engine size and weight highly correlated?
    - "Will independent variables create regression problems?"
- Why Can't We Directly Use Correlation?
    - correlation alone may not fully explain multicollinearity
    - variables may contain overlapping information

what is multicollinearity
- high correlation between independent variables.

| Purpose                           | Test Used |
| --------------------------------- | ---------- |
| check multicollinearity           | VIF        |

## question
- Independent variables:
    - engine size
    - weight

NULL HYPOTHESIS
- variables are not highly correlated.

Now what type of test we do
- checking multicollinearity
    - that's why we use VIF

what does VIF check
- how strongly one independent variable is explained by other independent variables.

what does VIF value mean
- larger VIF means stronger multicollinearity.

| VIF Value | Meaning                    |
| ----------| -------------------------- |
| 1         | no multicollinearity       |
| 1 - 5     | moderate correlation       |
| > 5       | high multicollinearity     |
| > 10      | serious problem            |

why large VIF is bad
- regression coefficients become unstable.
- difficult to identify feature importance.

important viva point
- VIF is used before regression analysis.
- it detects multicollinearity among independent variables.

```python
import pandas as pd

from sklearn.datasets import fetch_openml

from statsmodels.stats.outliers_influence import variance_inflation_factor

# Load Dataset
auto = fetch_openml(
    name="autoMpg",
    version=1,
    as_frame=True
)

df = auto.frame

# Select Independent Variables
X = df[["displacement", "weight"]]

# Remove Missing Values
X = X.dropna()

# Calculate VIF
vif_data = pd.DataFrame()

vif_data["Feature"] = X.columns

vif_data["VIF"] = [
    variance_inflation_factor(X.values, i)
    for i in range(X.shape[1])
]

print(vif_data)
```

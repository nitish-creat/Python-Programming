## p_value Test

why do we use p_value
- probability that result happened randomly.
- used to determine statistical significance.

why we use this
- example -> Did the new teaching method really improve scores?
    - "Is this increase statistically believable?"
- Why Can't We Directly Say 73 > 70 So Improvement Happened?
    - samples naturally vary
    - random fluctuations happen

what is p_value
- probability of getting the observed result if H₀ is true.

why do we calculate p_value
- How rare is this result if H₀ is true?

what is alpha
- maximum acceptable random error.
- significance level.
- usually 5% risk allowed (0.05)

why compare p_value with alpha
- to decide whether result is statistically significant or not.

| Condition | Decision          |
| --------- | ----------------- |
| p < alpha | Reject H₀         |
| p > alpha | Fail to Reject H₀ |

what does p < alpha mean
- result is unlikely to happen randomly.
- strong evidence against H₀.

what does p > alpha mean
- result may have happened randomly.
- not enough evidence against H₀.

what is statistically significant
- result is unlikely due to chance.

example
- p_value = 0.02
- alpha = 0.05

0.02 < 0.05
- Reject H₀
- result is statistically significant.

important viva point
- p_value tells probability of random occurrence.
- alpha tells acceptable random error limit.

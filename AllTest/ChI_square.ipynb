## Chi-Square Test

why do we use chi-square test
- to check relationship between categorical variables.
- used for frequency/count data.

why we use this
- example -> Does gender affect product preference?
    - "Are these categories related?"
- Why Can't We Directly Compare Frequencies?
    - frequencies naturally vary
    - random fluctuations happen

what is hypothesis
- assumption we test statistically.

| Data Type              | Test Used        |
| ---------------------- | ---------------- |
| category vs category   | chi-square test  |

## question
- Male Like = 40
- Female Like = 60
- Male Dislike = 30
- Female Dislike = 70
- Significance level = 0.05

NULL HYPOTHESIS
- H0 : Gender and product preference are independent.
    - no relationship exists.
- H1 : Gender and product preference are related.
    - relationship exists.

Now what type of test we do
- category vs category
    - that's why we use chi-square test

what does chi-square test check
- observed frequencies vs expected frequencies.

what does chi-square value mean
- how much observed data differs from expected data.
    - larger chi-square means stronger evidence.

what is p_value
- probability that observed difference happened randomly.

why do we use this
- How rare is this result if H₀ is true?

Why Compare p-value with Alpha?
- maximum acceptable random error
- 5% risk allowed (0.05)

| Condition | Decision          |
| --------- | ----------------- |
| p < alpha | Reject H₀         |
| p > alpha | Fail to Reject H₀ |

what does p < alpha mean
- relationship is statistically significant.
- variables are related.

what does p > alpha mean
- difference may have happened randomly.
- no strong evidence of relationship.

important viva point
- chi-square test is used for categorical variables.
- it checks association/relationship between categories.

```python
from scipy.stats import chi2_contingency
import pandas as pd
data = [
    [40,60],
    [30,70]
]

table = pd.DataFrame(
    data,
    columns=["Male","Female"],
    index=["Like","dislike"]
)
print(table)

chi2 , p_value, dof, expected = chi2_contingency(table)
print(p_value)
print(chi2)
print(dof)
print(expected)

if p_value < 0.05:

    print("\nReject Null Hypothesis")

    print(
        "Conclusion: "
        "Gender and product preference are significantly related."
    )

else:

    print("\nFail to Reject Null Hypothesis")

    print(
        "Conclusion: "
        "No significant relationship exists."
    )
```

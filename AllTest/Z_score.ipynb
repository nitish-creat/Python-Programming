##    Z_score Test
why do we use z test
- std known
- sample mean given
- sample large size is given 

why we use this
- example -> Did the new teaching method really improve scores? 
    - "Is this increase statistically believable?"
- Why Can't We Directly Say 73 > 70 So Improvement Happened?
    - samples naturally vary
    - random fluctuations happen

what is hypothesis
- assumption we test statistically.

| If Question Says   | Tail Type    |
| ------------------ | ------------ |
| increase / greater | right-tailed |
| decrease / lower   | left-tailed  |
| difference         | two-tailed   |



## question 
- Historical average score = 70
- Sample average = 73
- Sample size = 50
- Population standard deviation = 10
- Significance level = 0.05

NULL HYPOTHESIS
- H0 : mean = 70
    - new method did NOT improve scores.
- H1 : mean > 70
    - new method improved scores.

Now what type of test we do
- sample mean > pop mean
    - that's why we use right-tailed test 
    - for that we have to use 1 - norm.cdf(Z-score)
    - for left-tailed test we have to use only norm.cdf(Z-score) because cdf calculate from left to z_score

what does formual mean
- How far is sample mean from population mean?
    - in terms of STD

what z_score mean
- Sample mean is 2.12 standard deviations above population mean.
    - large z_score means stronger evidence
what is p_value
- probability that result happened randomly.
why do we use this
- How rare is this result if H₀ is true?

Why Compare p-value with Alpha?
- maximum acceptable random error
- 5% risk allowed (0.05)

| Condition | Decision          |
| --------- | ----------------- |
| p < alpha | Reject H₀         |
| p > alpha | Fail to Reject H₀ |



```python
import numpy as np
from scipy.stats import norm
pop_mean = 70
sample_mean = 73
n = 50
std = 10
a = sample_mean - pop_mean
b= std / np.sqrt(n)

z_score = a/b
print(f"z_score: {z_score:.2f}")
p_value = 1 - norm.cdf(z_score)
print(f"P_value : {p_value:.2f}")
if(p_value < 0.05): 
    print("reject the Null hyposthesis" \
    "\nso the new method imporoved score")
else:
    print("Failed to reject null hypo" \
    "\nnew method is not improving score")
```

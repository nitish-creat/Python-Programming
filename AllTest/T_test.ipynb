## T-test
why we use it 

| Condition            | Present? |
| -------------------- | -------- |
| comparing means      | ✅        |
| population std known | ❌        |
| small sample size    | ✅        |

type of t_test

| Type        | Use                      |
| ----------- | ------------------------ |
| One-sample  | one group vs population  |
| Independent | two separate groups      |
| Paired      | before-after same people |



## Question
- Technique A vs Technique B
- 10 students in each group
- Compare exam scores
- Population standard deviation NOT given

what question ask 
- Is Technique B really better than Technique A?
- or Did the difference happen randomly?

why can't we direclty compare the means like
- tech A means is 69
- tech B mean is 76
by this data we clearly see that b is good
- but Is this difference statistically significant?
- random sampling caused difference.

hence we are using independent two sample t-test(two seperate group)


given data:
- tech a -> student 10
- tech b -> student 10

hypothesis
- H0: mean of a == mean of b
    - both tech is same
- H1: mean of b > mean of a
    - tech b is good (what question wants !!)

Sample dataset

A = [65, 70, 68, 72, 66, 69, 71, 67, 73, 70]

B = [75, 78, 80, 74, 77, 79, 81, 76, 78, 80]

What Does ttest_ind() Do?
- it compare mean of b and a
- while considering variation and sample size

What is t-statistic?
- how far apart means are.
- large t_static means stronger evidence

| Condition | Decision          |
| --------- | ----------------- |
| p < alpha | Reject H₀         |
| p > alpha | Fail to Reject H₀ |



```python
import pandas as pd
from scipy.stats import ttest_ind
data = {
    "Technique_A": [65, 70, 68, 72, 66, 69, 71, 67, 73, 70],

    "Technique_B": [75, 78, 80, 74, 77, 79, 81, 76, 78, 80]
}

df = pd.DataFrame(data)
A = df["Technique_A"]
B = df["Technique_B"]

t_test , p_value = ttest_ind(B,A)
print(t_test)
print(p_value)
if(p_value <0.05):
    print("rejct Null hypo")
    print("tech b is improving the score")
else:
    print('fail to rejct null hypothesis')
```

## Shapiro-Wilk Test

why do we use shapiro-wilk test
- to check whether data follows normal distribution.
- used before parametric statistical tests.

why we use this
- example -> Are exam scores normally distributed?
    - "Can we safely use parametric tests?"
- Why Can't We Directly Look at Data?
    - visual inspection may be misleading
    - random variation happens

what is hypothesis
- assumption we test statistically.

| Purpose                  | Test Used           |
| ------------------------ | ------------------- |
| check normality of data  | shapiro-wilk test   |

## question
- Exam scores of 30 students
- Significance level = 0.05

NULL HYPOTHESIS
- H0 : Data is normally distributed.
- H1 : Data is not normally distributed.

Now what type of test we do
- checking normality
    - that's why we use shapiro-wilk test

what does shapiro test check
- how close data is to normal distribution.

what does test statistic mean
- measures similarity between data and normal distribution.
    - larger statistic means data is more normal.

what is p_value
- probability that deviation from normality happened randomly.

why do we use this
- How rare is this result if H₀ is true?

Why Compare p-value with Alpha?
- maximum acceptable random error
- 5% risk allowed (0.05)

| Condition | Decision          |
| --------- | ----------------- |
| p > alpha | Fail to Reject H₀ |
| p < alpha | Reject H₀         |

what does p > alpha mean
- data is normally distributed.
- no strong evidence against normality.

what does p < alpha mean
- data is not normally distributed.
- strong evidence against normality.

important viva point
- shapiro-wilk test checks normality of data.
- commonly used before t-test, ANOVA, and regression.

```python
from scipy.stats import shapiro

scores = [
    78, 82, 85, 88, 90,
    76, 79, 84, 91, 87,
    83, 80, 77, 86, 89,
    92, 81, 75, 88, 90,
    79, 84, 85, 87, 91,
    82, 80, 78, 86, 88
]

# Shapiro Test
stat, p_value = shapiro(scores)

print("Statistic:", stat)

print("P-value:", p_value)

# Significance Level
alpha = 0.05

# Decision
if p_value > alpha:

    print("Fail to Reject Null Hypothesis")

    print(
        "Conclusion: "
        "Data is normally distributed."
    )

else:

    print("Reject Null Hypothesis")

    print(
        "Conclusion: "
        "Data is not normally distributed."
    )
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
f = sns.load_dataset("flights")
print(f.info())

N = 5
men_means = (20, 35, 30, 35, 27)
women_means = (25, 32, 34, 20, 25)
men_std = (2, 3, 4, 1, 2)
women_std = (3, 5, 2, 3, 3)

ind = np.arange(N)
width = 0.35

p1 = plt.bar(ind, men_means, width, yerr=men_std)
p2 = plt.bar(ind, women_means, width,
             bottom=men_means, yerr=women_std)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))

plt.show()

plt.figure(figsize=(10,5))
f.boxplot(column="passengers", by="month")
plt.title("Boxplot of passengers by year")
plt.suptitle("")
plt.show()
```

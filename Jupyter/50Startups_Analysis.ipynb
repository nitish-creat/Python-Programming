```python
# Data Visualization: Distributions and Relationships
if df is not None:
    # 1. Histograms and Distribution Analysis
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    fig.suptitle('Distribution Analysis of Numerical Variables', fontsize=16, fontweight='bold')
    
    numerical_cols_list = list(numerical_cols)
    for idx, col in enumerate(numerical_cols_list[:4]):
        ax = axes[idx // 2, idx % 2]
        ax.hist(df[col], bins=20, edgecolor='black', color='skyblue', alpha=0.7)
        ax.set_title(f'{col} Distribution', fontweight='bold')
        ax.set_xlabel(col)
        ax.set_ylabel('Frequency')
        ax.grid(axis='y', alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # 2. Box Plots for Outlier Detection
    fig, axes = plt.subplots(1, len(numerical_cols_list), figsize=(16, 5))
    fig.suptitle('Box Plots - Outlier Detection', fontsize=16, fontweight='bold')
    
    for idx, col in enumerate(numerical_cols_list):
        ax = axes[idx] if len(numerical_cols_list) > 1 else axes
        box_plot = ax.boxplot(df[col], vert=True, patch_artist=True)
        box_plot['boxes'][0].set_facecolor('lightblue')
        ax.set_title(f'{col}', fontweight='bold')
        ax.set_ylabel('Value')
        ax.grid(axis='y', alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # 3. Correlation Heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
                fmt='.3f', square=True, linewidths=1, cbar_kws={'label': 'Correlation'})
    plt.title('Correlation Matrix Heatmap', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    # 4. Q-Q Plots for Normality Assessment
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    fig.suptitle('Q-Q Plots - Normality Assessment', fontsize=16, fontweight='bold')
    
    for idx, col in enumerate(numerical_cols_list[:4]):
        ax = axes[idx // 2, idx % 2]
        stats.probplot(df[col], dist="norm", plot=ax)
        ax.set_title(f'{col} Q-Q Plot', fontweight='bold')
        ax.grid(alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print("Normality Assessment (Shapiro-Wilk Test):")
    print("-" * 60)
    for col in numerical_cols_list:
        stat, p_value = stats.shapiro(df[col])
        normality = "Normal" if p_value > 0.05 else "Non-Normal"
        print(f"{col:20} | p-value: {p_value:.6f} | {normality}")
```

### Mathematical Formulations:

**Mean:** $\bar{x} = \frac{1}{n}\sum_{i=1}^{n} x_i$

**Variance:** $\sigma^2 = \frac{1}{n}\sum_{i=1}^{n}(x_i - \bar{x})^2$

**Standard Deviation:** $\sigma = \sqrt{\sigma^2}$

**Skewness:** $\gamma_1 = \frac{E[(X-\mu)^3]}{\sigma^3}$

**Kurtosis:** $\gamma_2 = \frac{E[(X-\mu)^4]}{\sigma^4} - 3$

**Covariance:** $\text{Cov}(X,Y) = \frac{1}{n}\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})$

**Correlation Coefficient:** $r = \frac{\sum(x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum(x_i - \bar{x})^2 \sum(y_i - \bar{y})^2}}$

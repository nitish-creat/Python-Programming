````python
import pandas as pd
import numpy as np
from scipy.stats import norm
data = {
    "Student_ID": range(1, 26),
    "Math_Score": [
        72, 75, 68, 80, 74,
        70, 69, 77, 71, 73,
        76, 78, 77, 72, 74,
        75, 79, 76, 70, 73,
        74, 71, 68, 82, 79,
    ]
}

df = pd.DataFrame(data)
# print(df)
# print(df.head(10))
print(df.mean())
print(df.std())
````

````python
pop_mean = 70
pop_std = 10
sample_mean = df['Math_Score'].mean()
n = len(df)
a = sample_mean - pop_mean
b = pop_std / np.sqrt(n)

z_score = a/b
print(f"Z score: {z_score.round(2)}")
print(f"Sample Mean: {sample_mean}")

p_value = 1 - norm.cdf(z_score)
print(p_value)

if(p_value < 0.05):
    print("Reject Hypothesis h0 : mean the value is increase significanlty")
    print("The new method is really helpfull for improving the marks")
else:
    print("No the value is not increase significatly")
    print("Marks are not improving")
````

### Given Information
- Population Mean (μ) = 70
- Sample Mean (x̄) = 73
- Sample Size (n) = 50
- Population Standard Deviation (σ) = 10
- Significance Level (α) = 0.05
---
### Objective
#### Check whether the new teaching method increased the average math score.
---
### Test Used
- One-Sample Z-Test
- Right-Tailed Test
---
### Hypotheses
#### Null Hypothesis
```math
H_0 : \mu = 70
```
#### Alternative Hypothesis
```math
H_1 : \mu > 70
```

What norm.cdf() Does

cdf means:

Cumulative Distribution Function

It gives:

probability from the left side up to the Z-score.

````python
sam_mean = 73
m = 70
n =50
std = 10
alpha = 0.05
# m < sam_mean -> right tailed test
a = sam_mean - m
b = std / np.sqrt(n)

z_score = a/b

print(z_score)

p_value = 1- norm.cdf(z_score)      # 1-() it means we are calculating for the right tailed Z-test
print(p_value)

if(p_value < alpha):
    print("it reject the Null hypothesis")
    print("The new teaching method significantly increased scores.")
else:
    print("NUll Hypothesis " \
    "the new teaching method is not significaltly indease")
````

- it means we have to do two sample t-test

Type of Test

Since checking:

Technique B > Technique A

✅ Right-Tailed Test

Hypotheses
 Null Hypothesis (H₀)

H₀: μA = μB
 Meaning:
Both techniques perform equally.

---

Alternative Hypothesis (H₁)

H₁: μB > μA

 Meaning:
Technique B performs better than Technique A.

- Question	Alternative Hypothesis
- B better than A	μB > μA
- A better than B	μA > μB
- any difference	μA ≠ μB

````python
from scipy.stats import ttest_ind

data = {
    "Technique_A": [65, 70, 68, 72, 66, 69, 71, 67, 73, 70],

    "Technique_B": [75, 78, 80, 74, 77, 79, 81, 76, 78, 80]
}

df = pd.DataFrame(data)
A = df["Technique_A"]
B = df["Technique_B"]

t_test , p_value = ttest_ind(B,A)
print(t_test)
print(p_value)


if p_value < alpha:

    print("\nReject Null Hypothesis")

    print(
        "Conclusion: "
        "Technique B significantly improves scores."
    )

else:

    print("\nFail to Reject Null Hypothesis")

    print(
        "Conclusion: "
        "No significant difference found."
    )
````

Why use Shapiro-Wilk Test?

To check whether data follows a normal distribution before applying parametric statistical tests.

| Parameter   | Value             |
| ----------- | ----------------- |
| Variable    | Monthly Sales     |
| Sample Size | 25                |
| Objective   | Check Normality   |
| Test Used   | Shapiro-Wilk Test |

| p-value  | Conclusion              |
| -------- | ----------------------- |
| p > 0.05 | normal distribution     |
| p < 0.05 | not normal distribution |


```python
from scipy.stats import shapiro
sales_data = [
    21500, 21800, 22700, 21300, 22900,
    24000, 23200, 22600, 23400, 22800,
    22000, 23600, 24800, 23900, 25000,
    24500, 25600, 26400, 26000, 25900,
    26300, 27000, 27500, 26800, 27600
]

stat , p_value = shapiro(sales_data)
print(stat)
print(p_value)

if p_value > 0.05:
    print("Fail to rejct the null hypo" \
    "\nsales data is normally distributed")
else:
    print("reject the null hypo" \
    "\nsales data are not normally distributed")
```

- Question 7

| Variable 1 | Gender                     |
| ---------- | -------------------------- |
| Variable 2 | Product Preference         |
| Data Type  | Categorical vs Categorical |
| Objective  | Check Relationship         |
What the Question is Asking

Check whether:

Gender and product preference are related
### Chi-Square Test of Independence

Why?
- both variables are categorical
- checking relationship/association
### Hypothesis
H0 :Gender and product preference are independent
no relationship exists.

H1: Gender and product preference are related
relationship exists

```python
from scipy.stats import chi2_contingency
import pandas as pd
data = [
    [40,60],
    [30,70]
]

table = pd.DataFrame(
    data,
    columns=["Male","Female"],
    index=["Like","dislike"]
)
print(table)

chi2 , p_value, dof, expected = chi2_contingency(table)
print(p_value)
print(chi2)
print(dof)
print(expected)

if p_value < 0.05:

    print("\nReject Null Hypothesis")

    print(
        "Conclusion: "
        "Gender and product preference are significantly related."
    )

else:

    print("\nFail to Reject Null Hypothesis")

    print(
        "Conclusion: "
        "No significant relationship exists."
    )
```

```python
import math
from scipy.stats import norm

mu = 150        
x_bar = 145     
sigma = 12      
n = 40          
alpha = 0.01    

z = (x_bar - mu) / (sigma / math.sqrt(n))

z_critical = norm.ppf(alpha)

p_value = norm.cdf(z)

print("Z-statistic:", z)
print("Critical Z-value:", z_critical)
print("p-value:", p_value)

if z < z_critical:
    print("Reject H0: The drug significantly reduces blood pressure.")
else:
    print("Fail to reject H0: No significant reduction in blood pressure.")
```
